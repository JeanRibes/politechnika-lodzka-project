```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
#from google.colab.patches import cv2_imshow

def imshow(img):
    plt.imshow(img, cmap = plt.cm.gray)
    plt.show()
def hist(img):
    histogram, bin_edges = np.histogram(img, bins=256, range=(0, 256))
    plt.plot(bin_edges[0:-1], histogram)
    plt.show()
def hist_log(img):
    histogram, bin_edges = np.histogram(img, bins=256, range=(0, 256))
    plt.plot(bin_edges[0:-1], np.log(histogram+1))
    plt.show()
img = cv2.imread('labeling/source/1_0.png')
grayscaled = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imshow(grayscaled)
```

```python
thresh = cv2.adaptiveThreshold(grayscaled, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 51,0)
#thresh = cv2.bitwise_not(thresh)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(6,6))
thresh2 = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
#thresh = cv2.morphologyEx(thresh, cv2.MORPH_ERODE, np.ones((3,3), np.uint8))
imshow(thresh2)
```

```python
sure_bg = cv2.dilate(
    cv2.bitwise_not(thresh)
                     ,np.ones((3,3),np.uint8),iterations=3)
imshow(sure_bg)
```

```python
inv = cv2.bitwise_not(thresh)
dist_transform = cv2.distanceTransform(inv,cv2.DIST_L2,5)
imshow(dist_transform)
ret, sure_fg = cv2.threshold(dist_transform,0.5*dist_transform.max(),255,0)
sure_fg = np.uint8(sure_fg)
imshow(sure_fg)
```

```python
unknown = cv2.subtract(sure_bg,sure_fg)
imshow(unknown)
```

```python
# Marker labelling
ret, markers = cv2.connectedComponents(sure_fg)
# Add one to all labels so that sure background is not 0, but 1
markers = markers+1
# Now, mark the region of unknown with zero
markers[unknown==255] = 0
imshow(markers)
```

```python
img = cv2.cvtColor(thresh,cv2.COLOR_GRAY2RGB)
markers = cv2.watershed(img,markers)
img[markers == -1] = [255,16,16]
imshow(img)
```

```python
watershedded = markers.copy()
imshow(markers)
watershedded[markers == 1] = 255
imshow(watershedded)
```

```python
#cv2.cvtColor(watershedded,cv2.COLOR_BGR2GRAY)
w_m = cv2.morphologyEx(np.uint8(watershedded), cv2.MORPH_CLOSE, cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(6,6)))
imshow(w_m)
imshow(grayscaled)
```
